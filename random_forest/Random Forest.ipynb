# Random Forest

```python
# Loading libraries
import pandas as pd
import decision_tree as dt # local lib
import random_forest as rf
import graphviz
from anytree.dotexport import DotExporter
```

```python
# Creating tree from benchmark dataset
tree = dt.create_decision_tree('./datasets/benchmark.csv', ';')

# Displaying benchmark tree
graphviz.Source(''.join([line for line in DotExporter(tree)]))
```

```python
matrix = rf.test_forest(pd.read_csv('./datasets/benchmark.csv', ';'), [tree])

print(rf.get_pretty_confusion_matrix(matrix))
```
